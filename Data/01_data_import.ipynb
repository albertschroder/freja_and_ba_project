# Data collect

Connect and retrieve data from Wharton Research Data Services.

```python
from google.colab import userdata
import pandas as pd
import wrds
import os
```

Import WRDS credentials.

```python
wrds_username = userdata.get("MY_WRDS_USERNAME")
wrds_password = userdata.get("MY_WRDS_USERNAME")
```

## Connect to WRDS

Code below connects to WRDS API.

```python
wrds_database = wrds.Connection(
    wrds_username = wrds_username,
    autoconnect=True
  )
```

## Query data

SQL code used query data is in `query.sql`.




```python
# Open save SQL queyr
with open("query.sql", "r") as file:
  query = file.read()

data_raw = wrds_database.raw_sql(query)
```

## Save data

The downloaded data is saved below.

```python
data_raw.to_parquet("handoffs/data_raw.parquet", engine="pyarrow", compression="snappy")
```
